# Regression Linéaire

## Definition

La régression linéaire est une approche statistique pour modéliser la relation entre une variable dépendante et une ou plusieurs variables indépendantes. Dans le cas le plus simple, la régression linéaire tente de dessiner une ligne droite qui s’adapte le mieux aux données d’entrée.

La formule de la régression linéaire est la suivante:

$ y = a + bx + e $

- y est la variable dépendante (la variable que nous essayons de prédire ou d’estimer),
- a est l’ordonnée à l’origine (le point où la ligne de régression croise l’axe des y),
- b est le coefficient de la variable indépendante (la pente de la ligne),
- x est la variable indépendante (la variable que nous utilisons pour faire la prédiction),
- e est l’erreur de prédiction (la différence entre la valeur réelle et la valeur prédite).

## Exemple


### Creation du data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Création de données d'exemple
np.random.seed(0)
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.rand(100, 1)

# Affichage des données
plt.scatter(x,y,s=10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Entrainement du modele

```python
# Création d'un modèle de régression linéaire
reg = LinearRegression()

# Entraînement du modèle avec nos données
reg.fit(x, y)

# Affichage du coefficient directeur et de l'ordonnée à l'origine
print(f"Le coefficient directeur est {reg.coef_[0][0]}")
print(f"L'ordonnée à l'origine est {reg.intercept_[0]}")
```

### Prediction

```python
# Prédiction avec notre modèle entraîné
y_pred = reg.predict(x)

# Affichage de la ligne de régression
plt.scatter(x, y, s=10)
plt.plot(x, y_pred, color='r')
plt.show()
```

## Quand utiliser la regression linéaire


La régression linéaire est utilisée lorsque nous voulons prédire une variable continue (comme le prix d’une maison) en fonction d’une ou plusieurs variables (comme la superficie de la maison et le nombre de chambres). Elle est également utile pour comprendre l’importance relative des variables indépendantes, c’est-à-dire qu’elle peut nous aider à comprendre quelle variable a plus d’impact sur la variable dépendante.

La régression linéaire est appropriée lorsque la relation entre les variables est linéaire, et les données sont bien distribuées avec une variance constante. Si ces conditions ne sont pas remplies, d’autres types de régression ou de modélisation seront plus approprié.
